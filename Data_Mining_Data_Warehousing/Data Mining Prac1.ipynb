```python
#Set A: Q1: Python program to prepare a scatter plot using Iris Dataset
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()

x = iris.data[:,0] #sepal length
y = iris.data[:,1] #sepal width

plt.scatter(x,y)
#label the axis
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

#Title of the plot
plt.title("Scatter Plot of Sepal Length vs Sepal Width (Iris Dataset)")
#display the plot
plt.show()
```

```python
#Set A: q2: program to prepare a scatter plot using the Forge dataset

import matplotlib.pyplot as plt
#forge-like dataset (2 features, 2 classes
X = [
    [1,2], [2,3], [3,1], [4,2],
    [5,2], [6,4], [7,3], [8,4]
]

y = [0,0,0,0,1,1,1,1]

#separate features
x1 = [point[0] for point in X]
x2 = [point[1] for point in X]

#create scatter plot
plt.scatter(x1, x2, c=y)
#label axis
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter plot of Forge Dataset")
plt.show()
```

```python
#Set: Bq1: program to find all null values in a dataset
import pandas as pd
data = pd.DataFrame({
    "Age":[21, None, 23, None],
    "Marks":[85,90, None, 88]
})
print(data)
print("\nNull values in each column:")
print("\n",data.isnull())
print("\n",data.isnull().sum())
```

```python
#Set B: q2: program to remove all null values in a given dataset
import pandas as pd
data = pd.DataFrame({
    "Age":[21, None, 23, None,20,25],
    "Marks":[85,90, None, 88,89,97]
})
print(data)
print("\nAfter removing null values the data is:\n")
print(data.dropna())
```
