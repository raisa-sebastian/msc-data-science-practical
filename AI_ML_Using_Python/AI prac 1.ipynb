```python
#Program 6: Learning agent that: Takes feeback (reward/penalty). Updates future decisions. Displays improvement over iteration.
import random
actions =["Action A","Action B"]
#initialize action values (learning memory)
action_values = [0,0]#index 0 for A, 1 for B
iterations = 10 #number the learning iterations
print("Learning agent stimulation")

for i in range(iterations):
    print(f"Iteration {i+1}")

    #agent selects action with highest value
    if action_values[0]>action_values[1]:
        action_index = 0
    elif action_values[0]<action_values[1]:
        action_index = 1
    else:
        action_index = random.randint(0,1)

action = actions[action_index]
print("Chosen Action:",action)

#environment gives feedback (stimulated)
#aassume action A is better than action B
if action == "Action A":
    reward = 1
else:
    reward = -1
print("Feedback:", "Reward" if reward == 1 else "Penalty")

#update learning (action values)
action_values[action_index] += reward

print("Updated Action Values: ",action_values)
print("-"*40)

print("Final learned action preference:")
print("Action A : ", action_values[0])
print("Action B : ", action_values[1])
```

```python
#Program 7:PEAS decription for Self-driving car
performance = [
    "Safety",
    "Minimum accidents",
    "Traffic rules",
    "Fuel Efficiency"
]
environment = [
    "Roads",
    "Traffic",
    "Pedestrains",
    "Weather",
]
actuators = [
    "Steering",
    "Accelerator",
    "Brakes",
    "Horn"
]
sensors = [
    "Cameras",
    "GPS",
    "LiDAR",
    "Speed Sensor"
]

print("PEAS decription for Self-driving car")
print("Performance Measures")
for item in performance:
    print(" -",item)

print("Environment Measures")
for item in environment:
    print(" -", item)

print("Actuators Measures")
for item in actuators:
    print(" -",item)

print("Sensors Measures")
for item in sensors:
    print(" -",item)
    
```

```python
#Program 6: Learning agent that: Takes feeback (reward/penalty). Updates future decisions. Displays improvement over iteration.
# using sleep, play and study as actions instead.
actions = ["Sleep", "Play", "Study"]
values = [0, 0, 1]   # initial value for each action
iterations = 5

print("Initial values:", values)
print()

for i in range(iterations):
    print("Iteration:", i + 1)
    
    # Choose action with highest value
    best = values.index(max(values))
    action = actions[best]

    print("Chosen Action:", action)
    
    # Feedback
    if action == "Study":
        values[best] += 1   # reward
    elif action == "Play":
        values[best] -= 1   # penalty
    # Sleep gives 0, no change needed

    print("Updated Values:", values)
    print()

# Final preferred action
final_best = values.index(max(values))
print("Final preferred action:", actions[final_best])
```
